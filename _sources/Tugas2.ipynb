# Matrik



## Definisi Matrik
Matriks adalah susunan bilangan, simbol, atau ekspresi yang disusun dalam bentuk baris dan kolom yang membentuk suatu persegi panjang. Setiap elemen dalam matriks ditempatkan pada posisi tertentu yang ditentukan oleh indeks baris dan kolomnya.

Secara umum, sebuah matriks dengan ukuran 𝑚 × 𝑛 m×n memiliki  𝑚 m baris dan 𝑛 n kolom, ditulis sebagai:

image.png

## Jenis-jenis Matriks:
1. Matriks Persegi: Matriks dengan jumlah baris dan kolom yang sama (𝑚 = n).
2. Matriks Nol: Semua elemen dalam matriks bernilai nol.
3. Matriks Identitas: Matriks persegi dengan elemen diagonal utama bernilai 1 dan elemen lainnya nol.
4. Matriks Diagonal: Matriks persegi di mana semua elemen di luar diagonal utama bernilai nol.
5. Matriks Segitiga: Matriks persegi yang hanya memiliki elemen di atas atau di bawah diagonal utama.
6. Matriks Transpos: Hasil dari menukar baris menjadi kolom dalam suatu matriks.

Matriks banyak digunakan dalam berbagai bidang, termasuk matematika, komputer, fisika, dan kecerdasan buatan, terutama dalam operasi aljabar linear.

## Oprasi Aritmatika Matrik
Operasi aritmatika matriks adalah serangkaian operasi matematika yang dilakukan pada matriks, seperti penjumlahan, pengurangan, perkalian, dan operasi lainnya yang mengikuti aturan tertentu. Operasi ini digunakan dalam berbagai bidang seperti matematika, fisika, ilmu komputer, dan kecerdasan buatan.

### Penjumlahan
#### Konsep Penjumlahan Matrik
Penjumlahan matriks adalah operasi yang dilakukan dengan menjumlahkan elemen-elemen yang bersesuaian dari dua matriks. Operasi ini hanya dapat dilakukan jika kedua matriks memiliki ukuran yang sama, yaitu memiliki jumlah baris dan kolom yang sama.

`

```python
import numpy as np
import sympy as sy

# Definisi matriks dengan SymPy
a_sy = sy.Matrix([[2, 5, 9], [1, 5, 7], [4, 0, 6]])
b_sy = sy.Matrix([[-3, 7, 8], [1, -4, 6], [5, 8, 0]])

# Penjumlahan matriks
sum_matrix = a_sy + b_sy
print("\nPenjumlahan Matriks:")
for i in range(sum_matrix.rows):
    print(f"{a_sy.row(i)} + {b_sy.row(i)} = {sum_matrix.row(i)}")
```

### Perkalian Matrik
#### Konsep Perkalian Matrik
Perkalian matriks adalah operasi aritmatika pada matriks yang dilakukan dengan menjumlahkan hasil kali elemen-elemen dari baris matriks pertama dengan elemen-elemen dari kolom matriks kedua. Operasi ini banyak digunakan dalam berbagai bidang seperti grafika komputer, kecerdasan buatan, dan pemodelan sistem.

```python
import numpy as np

# Langkah 1: Definisikan Matriks A dan B
# Pastikan jumlah kolom A sama dengan jumlah baris B
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Langkah 2: Lakukan Perkalian Matriks
# Menggunakan np.dot(A, B) atau operator @ untuk perkalian matriks
C = np.dot(A, B)

# Langkah 3: Tampilkan Matriks A dan B
print("Matriks A:")
print(A)

print("\nMatriks B:")
print(B)

# Langkah 4: Menjelaskan Perhitungan Setiap Elemen dalam Matriks Hasil
C_manual = np.zeros((A.shape[0], B.shape[1]))  # Matriks kosong untuk hasil manual

# Perhitungan manual elemen per elemen
C_manual[0, 0] = (A[0, 0] * B[0, 0]) + (A[0, 1] * B[1, 0])  # C[0,0]
C_manual[0, 1] = (A[0, 0] * B[0, 1]) + (A[0, 1] * B[1, 1])  # C[0,1]
C_manual[1, 0] = (A[1, 0] * B[0, 0]) + (A[1, 1] * B[1, 0])  # C[1,0]
C_manual[1, 1] = (A[1, 0] * B[0, 1]) + (A[1, 1] * B[1, 1])  # C[1,1]

# Langkah 5: Tampilkan Hasil Perkalian Matriks
print("\nHasil Perkalian Matriks A x B:")
print(C)

# Langkah 6: Menampilkan Perhitungan Manual untuk Memahami Cara Kerja
print("\nPerhitungan Manual untuk Setiap Elemen Matriks C:")
print(f"C[0,0] = (1×5) + (2×7) = {int(C_manual[0,0])}")
print(f"C[0,1] = (1×6) + (2×8) = {int(C_manual[0,1])}")
print(f"C[1,0] = (3×5) + (4×7) = {int(C_manual[1,0])}")
print(f"C[1,1] = (3×6) + (4×8) = {int(C_manual[1,1])}")
```
